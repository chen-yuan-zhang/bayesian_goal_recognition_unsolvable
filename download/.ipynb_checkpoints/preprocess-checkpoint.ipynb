```python
import json

filename = 'pilot_data.json'

with open(filename) as f:
    datas = json.load(f)

gr_instances = [i for i in range(1001, 1037)]
output = []

for data in datas:
    
    gr_data = [] 
    for item in data['content'].strip().split('\n'):

        info = item.strip().split('","')
        

        if info[3] == 'external-html-keyboard-response':
            id = info[7]
            tmp = {}
            print(info)
            print(info[0])
            if not info[0][1:]:
                input()
                continue
            rts = [int(t) for t in info[0][1:].split(',')]
            keys = [key for key in info[8].split(',')]
            print(info[9])
            if len(info[9].split('_')) == 4: # instance_gr_json_r.html...
                for index in range(len(keys)):
                    if keys[index] == '66':
                        keys[index] = 'A'
                    elif keys[index] == '65':
                        keys[index] = 'B'
            else:
                for index in range(len(keys)):
                    if keys[index] == '65':
                        keys[index] = 'A'
                    elif keys[index] == '66':
                        keys[index] = 'B'
            tmp['responses'] = [(key, rt) for key, rt in zip(rts, keys)]
            tmp['instance'] = info[9].split('=')[-2]
            

            if int(tmp['instance']) in gr_instances:
                gr_data.append(tmp)
    
    output.append({'id':id,'data':gr_data})

# print(output)
with open('gr_data.json', 'w') as outfile:
    json.dump(output, outfile)
```

```python
import json

filename = 'pilot_data.json'

with open(filename) as f:
    datas = json.load(f)

ps_instances = [i for i in range(20001, 20013)]
output = []

for data in datas:
    
    gr_data = [] 
    for item in data['content'].strip().split('\n'):

        info = item.strip().split('","')
        

        if info[3] == 'external-html-keyboard-response':
            id = info[7]
            tmp = {}
            print(info)
            print(info[0])
            if not info[0][1:]:
                input()
                continue
            rts = [int(t) for t in info[0][1:].split(',')]
            keys = [key for key in info[8].split(',')]
            print(info[9])
            
            set_up_flag = info[9].split('=')[-1]
            
            if set_up_flag === '000':
                for index in range(len(keys)):
                    if keys[index] == '37':
                        keys[index] = 'left'
                    elif keys[index] == '38':
                        keys[index] = 'up'
                    elif keys[index] == '39':
                        keys[index] == 'right'
                    elif keys[index] == '40':
                        keys[index] == 'down'
            
            if set_up_flag === '001':
                for index in range(len(keys)):
                    if keys[index] == '37':
                        keys[index] = 'right'
                    elif keys[index] == '38':
                        keys[index] = 'up'
                    elif keys[index] == '39':
                        keys[index] == 'left'
                    elif keys[index] == '40':
                        keys[index] == 'down'
                        
            if set_up_flag === '010':
                for index in range(len(keys)):
                    if keys[index] == '37':
                        keys[index] = 'left'
                    elif keys[index] == '38':
                        keys[index] = 'down'
                    elif keys[index] == '39':
                        keys[index] == 'right'
                    elif keys[index] == '40':
                        keys[index] == 'up'
                        
            if set_up_flag === '011':
                for index in range(len(keys)):
                    if keys[index] == '37':
                        keys[index] = 'right'
                    elif keys[index] == '38':
                        keys[index] = 'down'
                    elif keys[index] == '39':
                        keys[index] == 'left'
                    elif keys[index] == '40':
                        keys[index] == 'up'
                        
            if set_up_flag === '100':
                for index in range(len(keys)):
                    if keys[index] == '37':
                        keys[index] = 'up'
                    elif keys[index] == '38':
                        keys[index] = 'left'
                    elif keys[index] == '39':
                        keys[index] == 'down'
                    elif keys[index] == '40':
                        keys[index] == 'right'
                        
            if set_up_flag === '101':
                for index in range(len(keys)):
                    if keys[index] == '37':
                        keys[index] = 'down'
                    elif keys[index] == '38':
                        keys[index] = 'left'
                    elif keys[index] == '39':
                        keys[index] == 'up'
                    elif keys[index] == '40':
                        keys[index] == 'right'
                        
            if set_up_flag === '110':
                for index in range(len(keys)):
                    if keys[index] == '37':
                        keys[index] = 'up'
                    elif keys[index] == '38':
                        keys[index] = 'right'
                    elif keys[index] == '39':
                        keys[index] == 'down'
                    elif keys[index] == '40':
                        keys[index] == 'left'
                        
            if set_up_flag === '111':
                for index in range(len(keys)):
                    if keys[index] == '37':
                        keys[index] = 'down'
                    elif keys[index] == '38':
                        keys[index] = 'right'
                    elif keys[index] == '39':
                        keys[index] == 'up'
                    elif keys[index] == '40':
                        keys[index] == 'left'
                

            tmp['responses'] = [(key, rt) for key, rt in zip(rts, keys)]
            tmp['instance'] = info[9].split('=')[-2]
            
                
            

            if int(tmp['instance']) in ps_instances:
                gr_data.append(tmp)
    
    output.append({'id':id,'data':gr_data})

# print(output)
with open('ps_data.json', 'w') as outfile:
    json.dump(output, outfile)
```

```python
instance_stat = {}
for subject in output:
    data = subject['data']
    for instance in data:
        instance_id = int(instance['instance'])
        if instance_id in instance_stat:
            instance_stat[instance_id].append(instance['responses'])
        else:
            instance_stat[instance_id] = [instance['responses']]
```

```python
from statistics import median
ps_stats = {}
for instance in ps_instances:
    ins_data = instance_stat[instance]
    ps_stats[instance] = {'N':0, 'notN':0}
    ps_stats[instance]['steps'] = []
    
    for item in ins_data:
        print(item)
        if item[-1][1] == '78':
            ps_stats[instance]['N'] += 1
        else:
            ps_stats[instance]['notN'] += 1
            
        ps_stats[instance]['steps'].append(len(item) -2)
        
    ps_stats[instance]['mid_step'] = median(ps_stats[instance]['steps'])
    input()
    
for i in ps_instances:
    print(i)
    print(ps_stats[i])
    print()
        
```

```python
unsol_instances = [20001,20004,20005,20008,20009,20010]
for subject in output:
    steps = []
    for instance in subject['data']:
        instance_id = int(instance['instance'])
        if instance_id in unsol_instances:
            steps.append(len(instance['responses'])-2)
                
    print('pid:',subject['id'])
    print('steps in unsol_instances:',steps)
    print()
```
